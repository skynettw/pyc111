```python
url = "https://www.taiwannews.com.tw/en/news/4724989"
import requests
html = requests.get(url).text
print(html)
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(html, "lxml")
print(soup)
```

```python
import operator
sel = "#main > section.mod_single-column > article > div > span > p"
data = soup.select(sel)
#print(data)
mydata = ""
for p in data:
    mydata = mydata + p.text.strip()
mydata = mydata.translate({ord(c):None for c in list("(),.“”")})
mydata = mydata.split()
word_freq = dict()
for word in mydata:
    if word not in word_freq:
        word_freq[word] = 1
    else:
        word_freq[word] += 1
ordered_freq = sorted(word_freq.items(), key=operator.itemgetter(1), reverse=True)
for w, c in ordered_freq:
    print(w, c)
```
